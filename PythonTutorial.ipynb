For multiple columns use *iloc* while acessing by index

```python
df.loc[:, ['rating', 'carbo']]
```

We can sort by one or more rows using *sort_values*

```python
df.sort_values(['rating', 'carbo'], ascending=False)
```

Pandas also supports chaining meaning that you can perform an operation and then another operation on the output of the first option. 

Here I am going to *groupby* the manufacturer *mfr*, get the mean of all cereals within each meanufacturer and then sort by carbohydrates (in ascending order)

```python
df.groupby('mfr').agg({'carbo': 'mean'}).sort_values('carbo')
```

## Exploratory Data Visualization

Now that we know a little about our data and once we have it in a clean and organized state, let's generate some visualizations.

Correlation is a good place to start understanding relationship between variables.

```python
corr = df.corr()
corr
```

But this isn't easy to instantly recognize what the big and small numbers are without plotting them. Let's try it with matplotlib.

```python
import matplotlib.pyplot as plt

%matplotlib inline

plt.matshow(corr)
```

Not ideal. We can tweak the plot to show exactly what we want, but there's an even easier way. Let's try it with seaborn, a library built on top of matplotlib.

```python
import seaborn as sns
sns.heatmap(corr)
```

Let's try a *distplot*, a distribution plot or a histogram of the ratings.

```python
sns.distplot(df['rating'])
```
